```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/rianreiss/data-analistics/main/prof-leonardo/Cuisine_rating.csv'
cuisine_rating = pd.read_csv(url)
```

# Conhecendo os dados

```python
cuisine_rating.head()
```

**_.info()_** - Uma **visão geral** da estrutura do DataFrame, incluindo os **tipos de dados** e a **quantidade de valores não-nulos** em cada coluna.

```python
cuisine_rating.info()
```

**_.value_counts()_** - Exibe a **frequência absoluta** dos dados. Quantas vezes cada valor aparece.

```python
cuisine_rating['Cuisines'].value_counts()
```

**_Filtro_** - Filtrando valor de uma coluna.

```python
cuisine_rating[cuisine_rating['Cuisines'] == 'Japanese']
```

**_Média Filtrada_** - Calculando a média de um conjunto de dados, filtrados por algum critério.

```python
cuisine_rating[cuisine_rating['Cuisines'] == 'Japanese']['Overall Rating'].mean()
```

# Gráficos

```python
cuisine_rating['Gender']
```

```python
gender_counts = cuisine_rating['Gender'].value_counts()

gender_counts.head()
```

```python
labels = gender_counts.index
sizes = gender_counts.values
colors = ['lightblue', 'pink']  # Cores para 'Male' e 'Female'

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Aspecto igual garante que o gráfico seja desenhado como um círculo.
plt.title('Distribuição de Gênero')
plt.show()
```

```python
cuisine_rating['Location'] = cuisine_rating['Location'].replace({'Central Park,ny': 'Central Park,NY'})
# tratando dados que podem gerar erros na analise
```

```python
city_counts = cuisine_rating['Location'].value_counts()

city_counts.info()
```

```python
city_counts = cuisine_rating['Location'].value_counts()

# Ordenar as cidades por quantidade de registros
city_counts = city_counts.sort_values(ascending=False)

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))  # Define o tamanho da figura
city_counts.plot(kind='bar', color='skyblue')
plt.title('Quantidade de Registros por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Quantidade de Registros')
plt.xticks(rotation=45, ha='right')  # Rotaciona os rótulos do eixo x para facilitar a leitura
plt.tight_layout()  # Ajusta o layout para evitar corte de elementos
plt.show()
```
