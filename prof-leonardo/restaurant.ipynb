```python
import numpy as np
import pandas as pd
```

```python
cuisine_rating = pd.read_csv('Cuisine_rating.csv')
```

# Conhecendo os dados

```python
cuisine_rating.head()
```

**_.info()_** - Uma **visão geral** da estrutura do DataFrame, incluindo os **tipos de dados** e a **quantidade de valores não-nulos** em cada coluna.

```python
cuisine_rating.info()
```

**_.value_counts()_** - Exibe a **frequência absoluta** dos dados. Quantas vezes cada valor aparece.

```python
cuisine_rating['Cuisines'].value_counts()
```

**_Filtro_** - Filtrando valor de uma coluna.

```python
cuisine_rating[cuisine_rating['Cuisines'] == 'Japanese']
```

**_Média Filtrada_** - Calculando a média de um conjunto de dados, filtrados por algum critério.

```python
cuisine_rating[cuisine_rating['Cuisines'] == 'Japanese']['Overall Rating'].mean()
```
